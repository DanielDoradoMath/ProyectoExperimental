Este cuaderno muestra los datos obtenidos sobre las muestras de MAT2. Estos involucraron tomar medidas bajo el ciclo 0 -> V max -> V min -> V max -> 0.

Los datos fueron tomados para voltajes máximos/mínimos de +/- 1 V y +/- 5 V. Adicionalmente, para cada uno de estos voltajes, se tomaron datos con 20, 40 y 50 puntos. No fue posible aumentar más allá de 50 puntos, puesto que al tomar más de 50 la fuente se quedaba en un valor del voltaje y no completaba el ciclo.

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_dict = {}

for root, dirs, files in os.walk('./Datos/'):
    for name in files:
        if '.txt' in name:
            if 'External_' in name:
                V = name[name.find('nal_')+len('nal_'):name.rfind('_')]
                n_pts = name[name.find('V_')+len('V_'):name.rfind('pts')]
                data_dict[('Ext', V, n_pts)] = pd.read_csv('./Datos/'+name, delimiter='\t', 
                                         names=['Voltage (V)', 'Current (A)', 'Resistance (Ohms)'])
            elif 'Internal_' in name:
                V = name[name.find('nal_')+len('nal_'):name.rfind('_')]
                n_pts = name[name.find('V_')+len('V_'):name.rfind('pts')]
                data_dict[('Int', V, n_pts)] = pd.read_csv('./Datos/'+name, delimiter='\t', 
                                         names=['Voltage (V)', 'Current (A)', 'Resistance (Ohms)'])
```

```python
data20 = data_dict[('Int', '1V', '20')]
data40 = data_dict[('Int', '1V', '40')]
data50 = data_dict[('Int', '1V', '50')]

plt.plot(data20['Voltage (V)'], data20['Current (A)']*1e+6, marker='o', ms=3, color = 'darkgreen', label = '20 puntos')
plt.plot(data40['Voltage (V)'], data40['Current (A)']*1e+6, marker='o', ms=3, color = 'turquoise', label = '40 puntos')
plt.plot(data50['Voltage (V)'], data50['Current (A)']*1e+6, marker='o', ms=3, color = 'palegreen', label = '50 puntos')


plt.xlabel('Voltaje (V)')
plt.ylabel('Corriente (µA)')

plt.legend()
plt.show()
```
